```python
path = r'C:\projects\barber\Сводный отчет по парикмахерским 2024.xlsx'
```

```python
df = pd.read_excel(path, sheet_name='КГКМ ВЗиС 2-2')
```

```python
del_col = df.loc[: ,'ИТОГО выручка'::]
```

```python
del_col
```

```python
df_new = df.drop(columns=del_col.columns)
```

```python
df_new
```

```python
del_str = df_new['услуги'].str.contains('ИТОГ', na=False)
```

```python
del_str
```

```python
df_new = df_new[~del_str]
```

```python
df_new.head()
```

```python
df_temp = df_new.T
df_temp.reset_index(inplace=True)
df_temp = df_temp.rename(columns={'index': 'Услуги'})
df_temp
```

```python
del_unnamed = df_temp['Услуги'].str.contains('Unnamed')
del_unnamed
```

```python
df_temp['Услуги_1'] = np.where(df_temp['Услуги'].str.contains('Unnamed'), np.nan, df_temp['Услуги'])
df_temp
```

```python
df_temp = df_temp.drop(columns='Услуги')
```

```python
df_temp = df_temp.rename(columns={'Услуги_1': 'Услуги'})
df_temp
```

```python
df_temp['Услуги'] = df_temp['Услуги'].fillna(method = 'ffill')
```

```python
df_temp
```

```python
mean = df_temp['Услуги']
df_temp = df_temp.drop('Услуги', axis=1)
df_temp.insert(0, 'Услуги', mean)
```
