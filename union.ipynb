```python
import pandas as pd
import numpy as np
```

```python
path = r'C:\projects\barber\barber\Сводный отчет по парикмахерским 2024.xlsx'
```

```python
df = pd.read_excel(path, sheet_name='КГКМ ВЗиС 2-2')
```

```python
df.head()
```

```python
del_col = df.loc[: ,'ИТОГО выручка'::]
```

```python
del_col.head()
```

```python
df_new = df.drop(columns=del_col.columns)
```

```python
df_new
```

```python
del_str = df_new['услуги'].str.contains('ИТОГ', na=False)
```

```python
df_new = df_new[~del_str]
```

```python
df_new.head()
```

```python
df_new['услуги'] = df_new['услуги'].fillna(value=0)
```

```python
df_new['услуги'] = pd.to_datetime(df_new['услуги'])
```

```python
df_new['услуги']
```

```python
df_temp = df_new.T
df_temp.reset_index(inplace=True)
df_temp = df_temp.rename(columns={'index': 'Услуги'})
df_temp
```

```python
df_temp['Услуги'] = np.where(df_temp['Услуги'].str.contains('Unnamed'), np.nan, df_temp['Услуги'])
df_temp
```

```python
df_temp['Услуги'] = df_temp['Услуги'].fillna(method='ffill')
```

```python
df_temp.head()
```

```python
df_temp.columns = df_temp.iloc[0]
df_temp = df_temp.drop(0)
```
